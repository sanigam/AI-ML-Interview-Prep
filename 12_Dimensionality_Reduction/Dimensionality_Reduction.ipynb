```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.manifold import TSNE
from sklearn.datasets import make_blobs
np.random.seed(42)
print('Libraries loaded!')
```

---
## 1. PCA from Scratch

```python
# Generate correlated 3D data
X, y = make_blobs(n_samples=300, n_features=3, centers=3, random_state=42)

# PCA from scratch
X_centered = X - X.mean(axis=0)
cov = np.cov(X_centered.T)
eigenvalues, eigenvectors = np.linalg.eigh(cov)
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

# Project to 2D
X_pca_scratch = X_centered @ eigenvectors[:, :2]

# Compare with sklearn
pca = PCA(n_components=2).fit(X)
X_pca_sklearn = pca.transform(X)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].scatter(X_pca_scratch[:,0], X_pca_scratch[:,1], c=y, cmap='viridis', s=15)
axes[0].set_title('PCA from Scratch')
axes[1].scatter(X_pca_sklearn[:,0], X_pca_sklearn[:,1], c=y, cmap='viridis', s=15)
axes[1].set_title('PCA with sklearn')
plt.tight_layout(); plt.show()
```

---
## 2. Explained Variance & Scree Plot

```python
# High-dimensional data
X_hd, y_hd = make_blobs(n_samples=200, n_features=10, centers=3, random_state=42)
pca_full = PCA().fit(X_hd)
explained = pca_full.explained_variance_ratio_

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].bar(range(1, 11), explained, color='steelblue')
axes[0].set_title('Explained Variance per Component')
axes[0].set_xlabel('Component')
axes[1].plot(range(1, 11), np.cumsum(explained), 'ro-')
axes[1].axhline(y=0.95, color='gray', linestyle='--', label='95% threshold')
axes[1].set_title('Cumulative Explained Variance')
axes[1].legend(); axes[1].set_xlabel('Components')
plt.tight_layout(); plt.show()
n_95 = np.argmax(np.cumsum(explained) >= 0.95) + 1
print(f'Components for 95% variance: {n_95} out of {len(explained)}')
```

---
## 3. PCA vs t-SNE

```python
X_vis, y_vis = make_blobs(n_samples=300, n_features=20, centers=5, random_state=42)

pca2d = PCA(n_components=2).fit_transform(X_vis)
tsne2d = TSNE(n_components=2, perplexity=30, random_state=42).fit_transform(X_vis)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].scatter(pca2d[:,0], pca2d[:,1], c=y_vis, cmap='tab10', s=15)
axes[0].set_title('PCA (linear, preserves global structure)')
axes[1].scatter(tsne2d[:,0], tsne2d[:,1], c=y_vis, cmap='tab10', s=15)
axes[1].set_title('t-SNE (nonlinear, preserves local structure)')
plt.tight_layout(); plt.show()
print('PCA: fast, deterministic, good for preprocessing')
print('t-SNE: slow, stochastic, great for visualization')
```

## Key Interview Takeaways

1. **PCA** — linear, fast; use for preprocessing and visualization
2. **t-SNE** — nonlinear; great for visualization but NOT for preprocessing
3. **Scree plot** — choose components capturing 95%+ variance
4. **Curse of dimensionality** — high dims make distance metrics unreliable
5. **LDA** — supervised dimensionality reduction that maximizes class separation
